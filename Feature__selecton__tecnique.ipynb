# Feature Selection-Dropping constant features

In this step we will be removing the features which have constant features which are actually not important for solving the problem statement

```python
#Import pandas to create Dataframe
import pandas as pd
data=pd.DataFrame({"A":[1,2,4,1,2,4],"B":[4,5,6,7,8,9],"C":[0,0,0,0,0,0],"D":[1,1,1,1,1,1]})
```

```python
data.head()
```

# Varience Threshold
Feature selector that removes all low-varience features .

This feature selection algorithim looks only at the feature(X),not the desired outputs(y) and can thus be used for unsupervized learning.in this example all are input features

threshold=0.0 means ,check the feature which having zero varience,remove this fetature and you can see the c and D column has the same value so varience  and s.d and mean almost 0

```python
##I t will zero varience  features
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
#can change the threshold value to 0.1,0.02,0.3......1.0,2.0........that is actually called quasi constant feature dropping
var_thres.fit(data)
```

```python
var_thres.get_support()
```

in the output----it means that in the data A and B are true they are not zero varience where Cand D are zero varience remove those feature .So we are pickup those value which are False

```python
data.columns[var_thres.get_support()]#print only the true feature which are not remove
```

```python
#after do this----take out all my constant column  or which all column get remove will get it over here after this
```

```python
constant_columns=[column for column in data.columns
                  if column not in data.columns[var_thres.get_support()]]
print(len(constant_columns))
```

```python
#iterate the column name
for feature in constant_columns:
    print(feature)
```

```python
data.drop(constant_columns,axis=1)
```
