# Normalization(min-max Normalization )
In this approach we scale the values of the featres to a common range, usually between 0 and 1. T

he formula for this is:

Xnorm=(X-Xmin)/Xmax-Xmin

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('wine.csv',header=None,usecols=[0,1,2])
df.columns=["Class", "Alcohol", "Malic.acid"] # change wine name to class 
df.head()
```

```python
df = df.drop(index=0).reset_index(drop=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
```

```python
scaling.fit_transform(df[['Alcohol','Malic.acid']])
```

so scale down between 0 to 1 of alcohol and malik acid
