```python
#Independent features and dependent features
X=df.iloc[:,2:]
y=df.iloc[:,1]
```

# HoldOut Validation Approach -Train And Test Split

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=100)
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
result=model.score(X_test,y_test)
print(result)
```

As I change my random state then the accuracy may high and low  and this may actually overfitting and underfitting also because we don't know based on random state how we are splitting the datset and getting the accuracy .In order to overcome this we use Kfold CV

# K Fold Cross Validation

```python
from sklearn.model_selection import KFold
model=DecisionTreeClassifier()
kfold_validation=KFold(10)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score #for the Kfold we use this  and pass it to our model and see the accuracy  we use this
results=cross_val_score(model,X,y,cv=kfold_validation)
print(results)
print(np.mean(results))# getting average of all splits
```

# Leave One Out Cross Validation(LOOCV)
provides train /test indices to split data in train /test sets .Each sample is used one as a test set(singleton) while the remaining sample from training set

```python
from sklearn.model_selection import LeaveOneOut
leave_validation=LeaveOneOut()
model=DecisionTreeClassifier()
scores=cross_val_score(model,X,y,cv=leave_validation)
print(scores)
print(np.mean(scores))
```

# Repeated Random Test- Train Splits
This technique is a hybrid of traditional train-test splitting and the k-fold cross-validation method. In this technique, we create random splits of the data in the training-test set manner and then repeat the process of splitting and evaluating the algorithm multiple times, just like the cross-validation method.

```python
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeClassifier()
Ssplit=ShuffleSplit(n_splits=10,test_size=0.2)
results=cross_val_score(model,X,y,cv=Ssplit)#cross_val_score do the shuffle
print(results)
print(np.mean(results))
```
