Mutual Information

MI Estimate mutual information for a discrete target variable.

Mutual information (MI) between two random variables is a non-negative value, which measures the dependency between the variables. It is equal to zero if and only if two random variables are independent, and higher values mean higher dependency.

The function relies on nonparametric methods based on entropy estimation from k-nearest neighbors distances.

Inshort

A quantity called mutual information measures the amount of information one can obtain from one random variable given another.

The mutual information between two random variables X and Y can be stated formally as follows:


I(X ; Y) = H(X) – H(X | Y) Where I(X ; Y) is the mutual information for X and Y, H(X) is the entropy for X and H(X | Y) is the conditional entropy for X given Y. The result has the units of bits.

# let's do mutual information classification-feature selection

```python
from sklearn.feature_selection import mutual_info_classif
#determine the mutual information by the fomula give in above
mutual_info=mutual_info_classif(X_train,y_train,random_state=42)
mutual_info
```

 here never get any negative value either give + 1 value or 0,if  getting a very very high number that basically means that particular feature is best feature or is the most dependent feature or can say that dependency on the target variable  is too much
 Here the highest value is 0.7077.. check which feature is and near to target variable

 so do this first convert this mutual info into series in the next step(cell)

```python
mutual_info_series=pd.Series(mutual_info)#converting into series
mutual_info_series.index=X_train.columns#assign  to as index
mutual_info_sorting=mutual_info_series.sort_values(ascending=False)
mutual_info_sorting
```

Flavanoids  has the highest dependency on the target feature 'Wine'

```python
#let's plot the ordered mutual_info values per feature
mutual_info_sorting.plot.bar(figsize=(20,8))
```

```python
#select the top 5 best features -classification examples
from sklearn.feature_selection import SelectKBest
sel_10_best=SelectKBest(mutual_info_classif,k=5)
sel_10_best.fit(X_train,y_train)
X_train.columns[sel_10_best.get_support()]
```
